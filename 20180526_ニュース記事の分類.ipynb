```python
# OSライブラリの読み込み
import os
# 正規表現ライブラリの読み込み
import re

# sklearnライブラリの読み込み
from sklearn import datasets, model_selection
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Janomeライブラリの読み込み
from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
from janome.tokenfilter import POSKeepFilter

# pandas
import pandas as pd

# numpy
import numpy as np
```

```python
# ニュースフォルダの指定
dirs = ['it-life-hack', 'movie-enter']

# 各フォルダのファイルを一つずつ読み込んで表示
for i, d in enumerate(dirs):
    # ファイルの取得
    files = os.listdir('./data/text/' + d)
    
    for file in files:
        # ファイルオープンと読み込み
        f = open('./data/text/' + d + '/' + file, 'r', encoding='utf-8')
        raw = f.read()
        
        # 正規表現で不要な文字列を除去して表示
        reg_raw = re.sub(r'[0-9a-zA-Z]', '', raw)
        reg_raw = reg_raw.replace('¥n', '')
        print(reg_raw)
        
        # ファイルクローズ
        f.close()
```

```python
# ニュースの単語とラベルを格納するリストを生成
x_ls = [] # 説明変数
y_ls = [] # 目的変数

tmp1 = []
tmp2 = ''

# 形態素解析オブジェクトの生成
tokenizer = Tokenizer()
# 名詞のみ抽出するフィルターを生成
token_filter = [POSKeepFilter(['名詞'])]
# フィルターを搭載した解析フレームワークを生成
analyzer = Analyzer([], tokenizer, token_filter)

# 各フォルダのファイルを一つずつ読み込み、処理してリストへ格納
for i, d in enumerate(dirs):
    # ファイルの取得
    files = os.listdir('./data/text/' + d)
    
    for file in files:
        # ファイルオープンと読み込み
        f = open('./data/text/' + d + '/' + file, 'r', encoding='utf-8')
        raw = f.read()
        
        # 正規表現で不要な文字列を除去して表示
        reg_raw = re.sub(r'[0-9a-zA-z]', '', raw)
        reg_raw = reg_raw.replace('¥n', '')
        
        # 形態素解析し、単語をリストに格納
        for token in analyzer.analyze(reg_raw):
            tmp1.append(token.surface)
            tmp2 = ' '.join(tmp1)
        x_ls.append(tmp2)
        tmp1 = []
        
        # ニュースのラベルをリストに格納
        y_ls.append(i)
        
        # ファイルクローズ
        f.close()
```

```python
# 説明変数と目的変数をnumpy配列に変換
x_array = np.array(x_ls)
y_array = np.array(y_ls)

# 単語の出現回数をカウント
cntvec = CountVectorizer()
x_cntvecs = cntvec.fit_transform(x_array)
x_cntarray = x_cntvecs.toarray()

# データフレーム形式で単語の出現回数を表示
pd.DataFrame(x_cntarray)
```

```python
# 単語とインデックスの表示
for k, v in sorted(cntvec.vocabulary_.items(), key=lambda x: x[1]):
    print(k, v)
```

```python
# 単語のTF-IDFを計算
tfidf_vec = TfidfVectorizer(use_idf=True)
x_tfidf_vecs = tfidf_vec.fit_transform(x_array)
x_tfidf_array = x_tfidf_vecs.toarray()

# データフレーム形式で単語の出現回数を表示
pd.DataFrame(x_tfidf_array)
```

```python
# データセットを訓練用とテスト用に分割
train_X, test_X, train_Y, test_Y = model_selection.train_test_split(x_tfidf_array, y_array, test_size=0.2)

# データの長さを確認
print(len(train_X))
print(len(test_X))
```

```python
# PyTorchライブラリの読み込み
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
# 訓練用のテンソル作成
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()

# テスト用のテンソル作成
test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()

# テンソルの件数を表示
print(train_X.shape)
print(train_Y.shape)
```

```python
# 説明変数と目的変数のテンソルをまとめる
train = TensorDataset(train_X, train_Y)

# 一つ目のテンソルを確認
print(train[0])

# ミニバッチに分ける
train_loader = DataLoader(train, batch_size=100, shuffle=True)
```

```python
# ネットワークの作成
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(20779, 256)
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, 256)
        self.fc4 = nn.Linear(256, 128)
        self.fc5 = nn.Linear(128, 128)
        self.fc6 = nn.Linear(128, 2)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = F.relu(self.fc5(x))
        x = self.fc6(x)
        return F.log_softmax(x)
    
# インスタンスの生成
model = Net()
```

```python
# 誤差関数のセット
criterion = nn.CrossEntropyLoss()

# 最適化関数のセット
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 学習開始
for epoch in range(1000):
    total_loss = 0
    # 分割したデータの取り出し
    for train_x, train_y in train_loader:
        train_x, train_y = Variable(train_x), Variable(train_y)
        # 勾配をリセット
        optimizer.zero_grad()
        # 順伝播の計算
        output = model(train_x)
        # 誤差の計算
        loss = criterion(output, train_y)
        # 逆伝播の計算
        loss.backward()
        # 重みの更新
        optimizer.step()
        # 誤差の累積
        total_loss += loss.data[0]
        
    # 累積誤差を100回ごとに表示
    if (epoch+1) % 100 == 0:
        print(epoch+1, total_loss)
```

```python
# 計算グラフの構築
test_x, test_y = Variable(test_X), Variable(test_Y)
# 出力が0と1のどちらか
result = torch.max(model(test_x).data, 1)[1]
# モデルの精度を計算
accuracy = sum(test_y.data.numpy() == result.numpy()) / len(test_y.data.numpy())

# モデルの精度を表示
accuracy
```
